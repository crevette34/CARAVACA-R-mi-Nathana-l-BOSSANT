Introduction :

Ce projet vise à collecter et analyser les données d’occupation des parkings voitures et vélos de Montpellier. Grâce aux données enregistrées au format JSON, nous générons des graphiques illustrant l’évolution du nombre de places libres, ainsi qu’une heatmap de corrélation permettant d’identifier les tendances et les interactions entre les différents parkings.

2. Analyse de la journée

Les courbes montrent une forte diminution du nombre de places disponibles entre 20 000 et 60 000 secondes, soit entre environ 4h30 et 16h40. Cela correspond aux heures d’activité en ville. On peut expliquer cette tendance par plusieurs facteurs :


L'analyse des données sur une journée entière montre clairement que l'occupation des parkings suit le rythme de la ville. On remarque un cycle où les places se remplissent et se libèrent en fonction des horaires de travail, des déplacements des habitants et des activités commerciales.


Matinée (6h - 9h) : C’est le moment où la ville commence à s’animer. Les parkings se remplissent petit à petit, surtout avec ceux qui partent travailler, mais aussi avec les premiers clients des magasins qui ouvrent. On voit un mélange entre ceux qui se garent juste pour très peu de temps, comme acheter quelque chose ou prendre un café en terrasse, et ceux qui laissent leur voiture toute la journée parce qu’ils travaillent.


Milieu de journée (9h - 14h) : Au fil de la journée, les parkings se remplissent à fond et restent souvent presque pleins. C’est parce que plusieurs choses se passent en même temps : ceux qui travaillent ont déjà garé leur voiture depuis le matin, et en plus, plein de gens arrivent pour faire du shopping ou aller à un rendez-vous. À ce moment-là, de nouvelles personnes cherchent aussi une place, ce qui peut provoquer le fait que les parkings soient très occupés , surtout dans les endroits très fréquentés comme près des magasins du Polygone.

Après-midi (14h - 17h) : Contrairement à ce qui se passe après avoir manger le repas du midi, l’occupation des parkings ne diminue pas immédiatement, mais reste encore assez élevée. Cette stabilité du taux d’occupation peut s'expliquer par plusieurs raisons. D’une part, les travailleurs de bureau continuent leur journée et gardent leur place de stationnement. D’autre part, l’après-midi voit souvent de nouvelles, en raison des activités commerciales qui continuent d'attirer du monde avec par exemple des sorties scolaires. Parfois, un pic peut également se produire en début de soirée, ce qui provoque l’arrivée de nouvelles personnes.

Fin de journée (17h - 19h) : À ce moment-là, les parkings commencent à se vider petit à petit, surtout parce que les gens quittent le boulot et rentrent chez eux. Mais ça dépend des jours et des endroits. Par exemple, près des gros magasins ou des centres de loisirs, il y a encore pas mal de voitures, car certaines personnes en profitent pour faire du shopping, manger un bout ou juste se détendre en sortant.

Soirée (19h - 22h) : Plus la nuit avance, plus les parkings se vident et retrouvent de la place, un peu comme le matin. C’est normal, la plupart des magasins ferment, les gens finissent leur journée et rentrent chez eux. Mais dans certains endroits, comme près des cinémas, des restaurants, il y a encore pas mal de voitures, quand les gens sortent pour profiter de la soirée.

2. Facteurs influençant ces variations d’occupation

Le taux d’occupation des parkings ne suit pas juste un rythme boulot/repos, c’est un peu plus complexe que ça.

- L’activité économique et commerciale : Le parking Polygone, qui est collé à un gros centre commercial, est très occupé presque toute la journée. Le matin et l’après-midi, il y a plein de gens qui viennent faire du shopping, ou manger. Par contre, le parking Circé, même s’il est aussi bien rempli, semble être plus utilisé par ceux qui travaillent proche ou qui prennent les transports en commun.

-  L’influence des transports en commun : Les parkings relais près des trams ou des arrêts de bus jouent aussi un rôle. Beaucoup de gens préfèrent y laisser leur voiture pour continuer en transport en commun. Du coup, ça change les moments où les parkings sont pleins et ça fait qu’ils restent remplis de façon plus régulière dans la journée.

-  Comme l’analyse a été faite en semaine, le remplissage des parkings dépend surtout des trajets boulot-maison et des courses du quotidien. Si ça avait été un week-end, on aurait sûrement vu plus de monde pour les sorties et le shopping, avec des parkings qui se vident plus tôt le soir.

Heat Map de corrélation des différents parkings :

Une heatmap de corrélation permet de visualiser les relations entre les taux d’occupation des parkings sur une période donnée. Elle met en évidence les dépendances entre différents parkings, aidant ainsi à identifier différents paramètres.



h1.png

De cette manière là on pourrait se dire qu’on ne comprend pas trop ce qu’il y a sur cette heatmap et il est difficile de repérer pour définir leur relations.

C’est pour cela qu’on va effectuer plusieurs “zoom” afin de pouvoir y voir des relations ou non entre certains parkings voiture et vélo.

Donc pour déjà comprendre ce qu’on veut définir on va le faire sur un toute petite partie pour simplement expliquer le fonctionnement de la heatmap comme défini auparavant.

La Heat Map obtenue montre des zones bien distinctes : le rouge intense correspond à une corrélation de 1, indiquant des parkings ayant un comportement similaire, tandis que le bleu profond représente une corrélation de -1, révélant des parkings dont l’occupation évolue de manière opposée. Les zones beiges, proches de 0, indiquent une absence de corrélation significative entre certains parkings.


h2.png

Louis blanc et Gambetta : Ces deux parkings semblent avoir une corrélation positive (carré rouge/beige). Cela signifie que lorsque l'un des deux est plein, l'autre l'est probablement aussi. Ils pourraient être situés dans des zones similaires ou desservir une même clientèle.

Donc pour savoir si c’est parce qu’il y a beaucoup de place disponible au même moment ou il y a une forte occupation au même moment, il faut pour cela regarder le graphique.

Pour savoir qu’est-ce-qui faut regarder on sait déjà que Gambetta est un parkings voiture et Louis Blanc un parkings Vélo. Tout d’abord, sur le graphique j’ai remarqué que le nombre de place parking voiture disponible de Gambetta diminuer, si il y une forte corrélation entre les deux on peut en déduire que les places de parkings de Louis Blanc diminuer aussi.

Pour cela on peut prendre des données géographiques afin d’expliquer ce phénomène, nous savons que que les parkings se situent tous les deux proches du centre de Montpellier.

On peut donc en déduire plusieurs choses. De nombreuses personnes se mettent dans les places de parkings dans la plage horaire de 14h-16h, cela peut s’expliquer que les gens ont envie d’aller faire les magasins au centre ville comme dans les galerie marchandes, mais aussi les gens qui retourne travailler après avoir manger ( vers 14h ), vienne pour se garer pour aller à leur travaille.
Prenons un exemple diamétralement opposé pour comprendre pourquoi une Heat Map est vraiment intéressante.

Afin d’expliquer un exemple contraire un va prendre le cas des parkings “”Médiathèque Emile Zola” et “Renouvier”

Médiathèque Émile Zola: Ce parking attire principalement des personnes souhaitant se rendre à la médiathèque, souvent pour des études ou des loisirs. Les horaires d'affluence sont susceptibles d'être différents de ceux d'un parking situé dans une zone commerciale.

Pour le Parking Renouvier j’ai remarqué que le nombre de places libres n'était jamais le même la plupart du temps (à voir sur le graphique), j’ai remarqué qu’au début il avait 7 places de libre sur 8 et puis on bouge entre 4, 5 ensuite 3. Alors que pour le parking de la Médiathèque Emile Zola il reste la plupart du temps à 13 places libres sur 14.

On peut donc voir que la Heat Map correspond bien, car les 2 parkings évoluent de manière totalement opposée. Un son nombre de place libre ne bouge pas vraiment alors que l’autre son nombre de place libre évolue constamment.

On peut donc déduire ses résultats par le fait que le Parkings Renouvier se situe dans une zone où il est un peu éloigné du centre ville et on peut y voir des long trottoirs dans lesquels des vélos peuvent circuler.
